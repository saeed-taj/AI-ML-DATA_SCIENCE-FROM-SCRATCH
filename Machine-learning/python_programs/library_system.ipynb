## Task 5 : Library System

building a small library system. Books are stored in a dictionary with book name as key
and number of copies as value:

```python
books = {
"Python Basics": 3,
"Data Science Handbook": 2,
"C++ Guide": 4,
"AI for Beginners": 1
}
```

1 . Allow a user to borrow a book (reduce copy count).

2. If a book is not available, show an error message.

I did both part in one question 

```python
book_to_borrow = input("Enter the book you want : ")

if book_to_borrow in books:
    if books[book_to_borrow] > 0:  # here checking that books are available or not
        books[book_to_borrow] -= 1  # reducing the boook by 1
        print(f"You have borrowed '{book_to_borrow}'.")
        print(f"Copies left: {books[book_to_borrow]}")
    
    else:
        # If a book is not available, show an error message as said in the question already did here
        print(f"{book_to_borrow} is currently not available or invalid input")

else:
    # If a book is not available, show an error message as said in the question 
    print(f"{book_to_borrow} does not exist in the library.")


print("Updated library:", books)
```


 Keep track of borrowed books per user using a dictionary ({user: [books]}).
 
 Print a final report of available books and borrowed books.

 Both parts above also all of the parts of the TASK 5 are included here 

```python

books = {
"Python Basics": 3,
"Data Science Handbook": 2,
"C++ Guide": 4,
"AI for Beginners": 1
}

borrowed_books = {}  # track books borrowed by each user as said to make a dictionary in the question

while True: # used while loop
    user = input("\nEnter your name (or type 'exit' to finish): ")
    if user == "exit":
        break

    book_to_borrow = input("Enter the book you want to borrow: ")

    if book_to_borrow in books:
        if books[book_to_borrow] > 0:
            
            books[book_to_borrow] -= 1
            
            print(f"You have borrowed '{book_to_borrow}'. Copies left: {books[book_to_borrow]}")

            # track borrowed books per user
            if user in borrowed_books:
                borrowed_books[user].append(book_to_borrow)
            
            else:
                borrowed_books[user] = [book_to_borrow]
        
        else:
            print(f" {book_to_borrow} is currently not available.")
    
    else:
        print(f"{book_to_borrow} does not exist in the library.")

# Final report
print("\n   Final Report   ")
print("Available books:")

for book, copies in books.items(): # for loop used to print the copies left herer
    print(f"{book}: {copies} copies left")

print("\nBorrowed books per user:")
for user, borrowed in borrowed_books.items():
    print(f"{user}: {borrowed}")
    nfgldkfngld
```

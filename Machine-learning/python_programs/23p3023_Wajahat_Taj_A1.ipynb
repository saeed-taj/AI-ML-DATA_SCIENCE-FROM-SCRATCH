## Task 1: Online Store Transactions

```python
transaction = [
    (101 , "Ali" , "Laptop" , 1 , 75000),
    (102 , "Sara" , "Headphones" , 2 , 3500),
    (103 , "Ali" , "Mouse" , 1 , 1200),
    (104 , "Zara" , "Laptop" , 1 , 73000),
    (105 , "Omar" , "Keyboard" , 3 , 2500),
    ]

    
```

Calculate the total bill of each transaction.

```python

for iterate in transaction:
    transaction_id = iterate[0]
    name = iterate[1]
    product_name = iterate[2]
    quantity = iterate[3]
    price = iterate[4]

    total_bill = price * quantity

    print(f"Transaction_ID : {transaction_id} : {name } bought {product_name} for Total Bill : {total_bill}" )
```

Store customer spending in a dictionary ({customer: total_spent}).

```python
Dictionary = {}


for item in transaction:
    customer = item[1]
    spended = item[4] * item[3] # total is summed in the spended by multiplying number of quantity
    
    if customer in Dictionary:
        Dictionary[customer] += spended # if there are single person borrowed multiple things 
    else:
        Dictionary[customer] = spended
            
print(Dictionary)            
```

Find the customer who spent the most.

```python
Info = {}

for t in transaction:
    total = t[4] * t[3]
    name = t[1]

    if name in Info:
        Info[name] += total
    else:
        Info[name] = total

maximum_spended = max(Info, key = Info.get)  
print( maximum_spended ,Info[maximum_spended])  


# max_spended = max(Info , key = Info.get)
# print(max_spended , Info(max_spended))
```

## Task 2: Student Performance Report

```python
marks = {
"Areeba": [85, 90, 92],
"Bilal": [70, 65, 80],
"Hina": [95, 88, 91],
"Usman": [60, 72, 68],
"Zoya": [88, 77, 85]
}

```

Calculate the average marks of each student.

```python


for name, scores in marks.items():
    # sum() will directly sums elements of given list and len() gives the length so average = sum of total / number of elements
    Average = sum(scores) / len(scores)
    print(f"{name} : {Average:.1f}") # to get float or decimal numbers by 1 number after decimal 
    # .1f is for 1 decimal place
```

Use if–else to classify students as "Pass" (avg ≥ 70) or "Fail".

```python
for name, scores in marks.items():
    Average = sum(scores) / len(scores)

    if Average >= 70.0 and Average <= 100:
        print(f"{name} : Pass")

    elif Average >= 1 and Average < 70:
        print(f"{name} : Fail")

    else:
        print("Invalid...")    

```

```python
All_averages = {}  # new dictionary to store the name and average of each student

for name, scores in marks.items():
    Average = sum(scores) / len(scores)
    All_averages[name] = Average  

Highest_result = max(All_averages , key=All_averages.get)  # calculating highest averaage
print(f"{Highest_result} Scored the Highest Average : {All_averages[Highest_result]:.1f} ")


All_averages[name] = Average

Highest = max(All_averages , key=All_averages.get)
```

[("Areeba", 89, "Pass"), ("Bilal", 71, "Pass"), ...]

```python
New_List = []

for name, scores in marks.items():
    Average = sum(scores) / len(scores)
    Result = "Pass" if Average >= 100 else "Fail"

    New_List.append((name, round(Average) , Result))

print(New_List)
```

## Task 3 : Text Analysis

Take the following text:

Python is simple yet powerful.
C++ is powerful but not as simple as Python.
Both languages are popular among programmers.

Convert the text to lowercase and split into words.

```python
Text = """Python is simple yet powerful.
C++ is powerful but not as simple as Python.
Both languages are popular among programmers."""

# Converted to lower case
Lower_Text = Text.lower()

# splitting into words as said 
Words = Lower_Text.split()

print(Words)

```

Use a set to find all unique words.

```python
Unique_Words = set(Words)
print(Unique_Words)
```

Count how many times each word appears (use a dictionary).  

```python
# as i converted the the text into lower case in the above code which was stored in Words i use that variable here
word_count = {} # dictionary as said

for word in Words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

for word, count in word_count.items():
    print(word, ":", count)
```

Find the most frequent word(s).

```python
# maximum frequency (the largest value)
max_count = max(word_count.values())   # e.g. 3


#  all words having that count
most_frequent_words = [w for w, c in word_count.items() if c == max_count]

print("Most frequent word(s):", most_frequent_words)
print("Frequency:", max_count)
```

```python
# two sentences initially
sentence_01 = "Python is simple yet powerful."
sentence_02 = "C++ is powerful but not as simple as Python."

              # lowercase and remove punctuation
import string

def clean(sentence):
    s = sentence.lower() # change the sentence to lower 
    for p in string.punctuation: 
        s = s.replace(p, "")
    return s.split()   # returns list of words

word_s1 = set(clean(sentence_01))   # convert to set for uniqueness
word_s2 = set(clean(sentence_02))

# intersection (common words)
common_words = word_s1 & word_s2   # or words entence_01.intersection(words2)

print("Words in both sentences:", common_words)
```

```python
expenses = [
{"name": "Ali", "items": [("Hotel", 5000), ("Food", 2000)]},
{"name": "Sara", "items": [("Travel", 3000), ("Food", 1500)]},
{"name": "Bilal", "items": [("Hotel", 5000), ("Shopping", 4000)]}
]
```

Calculate total expense of each person.

```python

for record in expenses:
    name = record["name"]
    items = record["items"]
    
    total = sum(amount for category, amount in items) # sum the second element (amount) of each tuple
    
    print(f"{name} : {total}")
```

Find the overall total expense.

```python
overall_total = 0  # initial value for total

for record in expenses:                   # go through each person
    items = record["items"]             
    total_for_person = sum(amount for i, amount in items)  # sum amounts
    overall_total += total_for_person     # add to the grand total

print("Overall total expense:", overall_total) 
```

```python

category_totals = {}  # dictionary to store total per category

for record in expenses:
    for category, amount in record["items"]:  # nested for loop to iterate through the items indide the list 
        if category in category_totals:
            category_totals[category] += amount
        else:
            category_totals[category] = amount

# Find the category with the maximum total
max_category = max(category_totals, key=category_totals.get)

print("Category totals:", category_totals)
print("Most expensive category:", max_category, "with total", category_totals[max_category])
```

## Task 5 : Library System

You are building a small library system. Books are stored in a dictionary with book name as key
and number of copies as value:

```python
books = {
"Python Basics": 3,
"Data Science Handbook": 2,
"C++ Guide": 4,
"AI for Beginners": 1
}
```

1 . Allow a user to borrow a book (reduce copy count).

2. If a book is not available, show an error message.

I did both part in one question 

```python

book_to_borrow = input("Enter the book you want : ")

if book_to_borrow in books:
    if books[book_to_borrow] > 0:  # here checking that books are available or not
        books[book_to_borrow] -= 1  # reducing the boook by 1
        print(f"You have borrowed '{book_to_borrow}'.")
        print(f"Copies left: {books[book_to_borrow]}")
    
    else:
        # If a book is not available, show an error message as said in the question already did here
        print(f"{book_to_borrow} is currently not available or invalid input")

else:
    # If a book is not available, show an error message as said in the question 
    print(f"{book_to_borrow} does not exist in the library.")


print("Updated library:", books)
```


 Keep track of borrowed books per user using a dictionary ({user: [books]}).
 
 Print a final report of available books and borrowed books.

 Both parts above also all of the parts of the TASK 5 are included here 

```python

books = {
"Python Basics": 3,
"Data Science Handbook": 2,
"C++ Guide": 4,
"AI for Beginners": 1
}

borrowed_books = {}  # track books borrowed by each user as said to make a dictionary in the question

while True: # used while loop
    user = input("\nEnter your name (or type 'exit' to finish): ")
    if user == "exit":
        break

    book_to_borrow = input("Enter the book you want to borrow: ")

    if book_to_borrow in books:
        if books[book_to_borrow] > 0:
            
            books[book_to_borrow] -= 1
            
            print(f"You have borrowed '{book_to_borrow}'. Copies left: {books[book_to_borrow]}")

            # track borrowed books per user
            if user in borrowed_books:
                borrowed_books[user].append(book_to_borrow)
            
            else:
                borrowed_books[user] = [book_to_borrow]
        
        else:
            print(f" {book_to_borrow} is currently not available.")
    
    else:
        print(f"{book_to_borrow} does not exist in the library.")

# Final report
print("\n   Final Report   ")
print("Available books:")

for book, copies in books.items(): # for loop used to print the copies left herer
    print(f"{book}: {copies} copies left")

print("\nBorrowed books per user:")
for user, borrowed in borrowed_books.items():
    print(f"{user}: {borrowed}")
```
